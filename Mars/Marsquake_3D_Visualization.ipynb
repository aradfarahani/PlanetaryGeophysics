# ü™ê Marsquake 3D Visualization

This notebook visualizes seismic activity on Mars using data from NASA‚Äôs InSight mission, overlaid on Mars topography.

## üîç Data Sources

- **Seismic Events:**  
  `events_extended_multiorigin_v14_2023-01-01.xml`  
  Parsed to extract Marsquake latitude, longitude, and magnitude.

- **Mars Topography:**  
  Created using:
  - `megt90n000eb.lbl`
  - `megt90n000eb.img`  
  Converted to a GeoTIFF file (`mars_mola_16ppd.tif`) using GDAL tools.

## üåç Visualization

- Elevation is rendered as a 3D globe using **Plotly**.
- Marsquakes are plotted as color-coded, size-scaled markers based on magnitude.
- The **InSight lander** location is marked at (135.6¬∞E, 4.5¬∞N).

The result is an interactive globe that can be rotated and zoomed to explore Mars‚Äô surface and seismic activity.


```python
from obspy.core.event import read_events
import pandas as pd

# Load Marsquake events from QuakeML XML file
catalog = read_events("events_extended_multiorigin_v14_2023-01-01.xml")

# Parse catalog into a DataFrame
data = {
    "time": [],
    "latitude": [],
    "longitude": [],
    "depth_km": [],
    "magnitude": []
}

for event in catalog:
    try:
        origin = event.preferred_origin() or event.origins[0]
        mag = event.preferred_magnitude() or event.magnitudes[0]
        data["time"].append(origin.time.datetime)
        data["latitude"].append(origin.latitude)
        data["longitude"].append(origin.longitude)
        data["depth_km"].append(origin.depth / 1000 if origin.depth else None)
        data["magnitude"].append(mag.mag)
    except Exception:
        continue

df = pd.DataFrame(data)
print(df.head())
```

# Mars Topography and Seismic Activity from InSight Mission

```python
import matplotlib.pyplot as plt
import numpy as np
import rasterio
import pandas as pd
from matplotlib_scalebar.scalebar import ScaleBar
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# === Load Mars elevation data ===
with rasterio.open("mars_mola_16ppd.tif") as src:
    elevation = src.read(1).astype(float)

# Clean no-data values
elevation[elevation > 10000] = np.nan

# === Generate coordinate grids for contours ===
lon_grid = np.linspace(-180, 180, elevation.shape[1])
lat_grid = np.linspace(-90, 90, elevation.shape[0])

# === Create figure and main axis ===
fig, ax = plt.subplots(figsize=(14, 8))

# === Plot base elevation map with Mars-style colormap ===
im = ax.imshow(elevation, cmap='gist_earth', extent=[-180, 180, -90, 90])
ax.axis('off')
ax.set_title("Mars MOLA Topography with Seismic Events", pad=20)

# === Add contour lines ===
levels = np.linspace(np.nanmin(elevation), np.nanmax(elevation), 40)
contours = ax.contour(
    lon_grid, lat_grid, elevation,
    levels=levels, colors='black', linewidths=0.4, alpha=0.2
)

# === Plot Marsquake data (assuming 'df' already exists) ===
sc = ax.scatter(
    df["longitude"], df["latitude"],
    c=df["magnitude"], cmap="Reds", alpha=0.8,
    edgecolors="black", linewidths=0.3,
    s=(df["magnitude"].fillna(1)**2.5)*5
)

# === Mark InSight Lander location ===
ax.scatter(135.6, 4.5, marker='*', s=120, c='white', edgecolors='black', linewidths=1, zorder=10)
ax.text(135.6 + 2, 4.5, "InSight", color='white', fontsize=9, weight='bold', va='center', zorder=10)

# === Add colorbars ===
# Vertical elevation colorbar to the left
cbar_left_ax = fig.add_axes([0.1, 0.2, 0.015, 0.6])
cb_left = fig.colorbar(im, cax=cbar_left_ax, orientation='vertical')
cb_left.set_label('Elevation (m)')

# Horizontal magnitude colorbar below
cbar_bottom_ax = fig.add_axes([0.3, 0.07, 0.4, 0.015])
cb_bottom = fig.colorbar(sc, cax=cbar_bottom_ax, orientation='horizontal')
cb_bottom.set_label('Magnitude')

# === Add a magnitude legend ===
legend_mags = [2.0, 3.0, 4.0]
for mag in legend_mags:
    ax.scatter([], [], s=(mag**2.5)*5, c='red', alpha=0.6, edgecolors='black',
               label=f'M {mag:.1f}')
ax.legend(scatterpoints=1, frameon=True, labelspacing=1, loc='lower left',
          title="Marsquake Magnitude")

# === Add a scale bar (1 degree ‚âà 59 km) ===
scalebar = ScaleBar(59_000, units='m', dimension='si-length', location='upper right',
                    scale_loc='bottom', length_fraction=0.1, color='white', box_color='black')
ax.add_artist(scalebar)

# === Add global inset map ===
inset_ax = inset_axes(ax, width="20%", height="20%", loc='upper left')
inset_ax.imshow(elevation, cmap='gist_earth', extent=[-180, 180, -90, 90])
inset_ax.scatter(135.6, 4.5, marker='*', s=40, c='white')
inset_ax.set_xticks([]); inset_ax.set_yticks([])
inset_ax.set_title("Global View", fontsize=8)

# === Final layout and save/display ===
plt.tight_layout(rect=[0.16, 0.1, 1, 0.92])   # leave space for elements
plt.savefig("mars_topography_quakes.png", dpi=300)
plt.show() 
```

## Load and process Mars elevation data

```python
import numpy as np
import rasterio

# Load Mars MOLA elevation data
with rasterio.open("mars_mola_16ppd.tif") as src:
    elevation = src.read(1).astype(float)
    elevation[elevation > 10000] = np.nan  # Clean no-data values

# Downsample elevation for performance
elevation = elevation[::10, ::10]

# Create latitude and longitude grids
lat_grid = np.linspace(-90, 90, elevation.shape[0])
lon_grid = np.linspace(-180, 180, elevation.shape[1])
lon, lat = np.meshgrid(lon_grid, lat_grid)

# Normalize elevation for coloring
elev_norm = np.nan_to_num(elevation)
elev_norm = (elev_norm - np.min(elev_norm)) / np.ptp(elev_norm)
```

## Plot interactive 3D Mars globe with quake points


$$
\begin{cases}
x = R \cdot \cos(\text{lat}) \cdot \cos(\text{lon}) \\
y = R \cdot \cos(\text{lat}) \cdot \sin(\text{lon}) \\
z = R \cdot \sin(\text{lat})
\end{cases}
$$

where lat and lon are in **radians**, and $R$ is Mars‚Äô radius (about 3389.5 km).


### Why this matters:

* Lat/Lon as flat x/y ignores Mars‚Äô curvature ‚Üí looks like a flat map.
* Converting to Cartesian places points on the spherical surface ‚Üí true globe visualization.
* You can then plot surface + points **around a sphere** that rotates naturally.

You can scale $R$ to 1 if you want a normalized sphere (unit sphere) and exaggerate elevation on top of that.


```python
import plotly.graph_objects as go

# Parameters for Mars globe
mars_radius = 1.0
elev_scale = 0.05

# Convert to radians
lat_rad = np.radians(lat)
lon_rad = np.radians(lon)

# Calculate radial distances
r = mars_radius + elev_norm * elev_scale

# Convert spherical to Cartesian coordinates for surface
x = r * np.cos(lat_rad) * np.cos(lon_rad)
y = r * np.cos(lat_rad) * np.sin(lon_rad)
z = r * np.sin(lat_rad)

# Surface plot
surface = go.Surface(
    x=x,
    y=y,
    z=z,
    surfacecolor=elev_norm,
    colorscale='Earth',
    showscale=False,
    lighting=dict(ambient=0.5, diffuse=0.5)
)

# Convert quake lat/lon to spherical Cartesian coordinates
quake_lat_rad = np.radians(df["latitude"])
quake_lon_rad = np.radians(df["longitude"])
quake_r = mars_radius + elev_scale
quake_x = quake_r * np.cos(quake_lat_rad) * np.cos(quake_lon_rad)
quake_y = quake_r * np.cos(quake_lat_rad) * np.sin(quake_lon_rad)
quake_z = quake_r * np.sin(quake_lat_rad)

# Quake scatter plot
quake_scatter = go.Scatter3d(
    x=quake_x,
    y=quake_y,
    z=quake_z,
    mode='markers',
    marker=dict(
        size=df["magnitude"] * 2,
        color=df["magnitude"],
        colorscale='Reds',
        colorbar=dict(
            title='Magnitude',
            thickness=20,
            len=0.6,
            y=0.5,
            x=1.05,
            outlinewidth=1,
            bordercolor='black'
        ),
        line=dict(width=0.5, color='black')
    ),
    name='Marsquakes'
)

# InSight lander coordinates
insight_lat = 4.5
insight_lon = 135.6
insight_lat_rad = np.radians(insight_lat)
insight_lon_rad = np.radians(insight_lon)
insight_r = mars_radius + elev_scale
insight_x = insight_r * np.cos(insight_lat_rad) * np.cos(insight_lon_rad)
insight_y = insight_r * np.cos(insight_lat_rad) * np.sin(insight_lon_rad)
insight_z = insight_r * np.sin(insight_lat_rad)

insight = go.Scatter3d(
    x=[insight_x],
    y=[insight_y],
    z=[insight_z],
    mode='markers+text',
    text=["InSight"],
    textposition='top center',
    marker=dict(symbol='cross', size=8, color='white', line=dict(color='black', width=1)),
    name='InSight Lander'
)

# Compose figure
fig = go.Figure(data=[surface, quake_scatter, insight])

fig.update_layout(
    title="3D Interactive Mars Globe with Marsquakes",
    scene=dict(
        xaxis=dict(showbackground=False, showticklabels=False, title=''),
        yaxis=dict(showbackground=False, showticklabels=False, title=''),
        zaxis=dict(showbackground=False, showticklabels=False, title=''),
        aspectmode='data'
    ),
    margin=dict(l=0, r=80, b=0, t=50),
)

fig.write_html("mars_globe.html")  
fig.show()
```
